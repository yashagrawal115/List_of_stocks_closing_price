```python
from pandas_datareader import data
import re
import timeit
```

## Import the data from the Yahoo Finance site

#### We have taken data of last 30 days closing price of listed stocks for current financial years.


```python
% timeit
stock_list = Stock_list['SYMBOL    '][:].values.tolist()

new_stock_list = []
for stock in stock_list:
  stock_s1= re.sub('$','.NS',stock)
  stock_s2 = re.sub('\s','',stock_s1)
  new_stock_list.append(stock_s2)

for stock in new_stock_list:
  fb = data.DataReader(new_stock_list, 'yahoo',start='10/17/2019', end='10/19/2019')
  fb[['Close']]
```

```python
fb['Close']
```

```python
stock_closing = fb['Close']
```

```python
stock_closing.to_csv('Stock_with_closing.csv')
```
